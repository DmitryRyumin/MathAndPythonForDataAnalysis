Подключаем необходимые библиотеки; сообщаем, что графики нужно строить прямо в Jupyter Notebook:

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
````

### В качестве непрерывного распределения выбираем хи-квадрат ($\chi^2$) распределение.

Данное распределение имеет один параметр $k$, который также называется степенью свободы (degrees of freedom, df). Математическое ожидание $\mathbf{E}X = k$ и дисперсия $\mathbf{D}X = 2k$. Для генерации выборки воспользуемся модулем ```stats``` библиотеки ```Scipy```. Объем выборки обозначим $n$.

````python
k = 5
n = 1000
chi = sts.chi2(k) # задаем распределение с параметром k
sample = chi.rvs(n) # генерируем выборку объемом n
#print(sample)
````

Построим теперь гистограмму выборки и теоретическую плотность распределения на одном графике:

````python
x = np.linspace(0, 30, 200)
plt.hist(sample, density=True, bins = 20, histtype='bar', ec='black')
plt.plot(x, chi.pdf(x))
plt.title('Гистограмма выборки и теоретическая плотность распределения')
plt.xlabel('$x$')
````

Из графика видно хорошее соответствие гистограммы выборки и теоретической плотности. Это связано с тем, что выборка достаточно велика.

### Распределение выборочного среднего

Фиксируем теперь значения объёма выборки n: (5, 10, 50).

#### Объём выборки $n=5$:

````python
k = 5
n = 5
chi = sts.chi2(k) # задаем распределение с параметром k
sample = chi.rvs(n) # генерируем выборку объемом n
#print(sample)
````

Генерируем выборку выборочных средних:

````python
sample_mean_arr = [] # создаем список для хранения выборочных средних
for i in range(1000): # цикл по выборкам
    sample = chi.rvs(n) # генерируем выборку объема n
    sample_mean = sum(sample) / n # вычисляем выборочное среднее как среднее арифметическое
    sample_mean_arr.append(sample_mean) # добавляем полученное значение в список

sample_mean_arr = np.array(sample_mean_arr) # переходим от списка к массиву numpy
````

Строим гистограмму и плотность нормального распределения. Параметры нормального распределения: $\mu = \mathbf{E}X = k, \sigma = \sqrt{\dfrac{\mathbf{D}X}{n}} = \sqrt{\frac{2k}{n}}$.

````python
plt.hist(sample_mean_arr, density=True, bins = 20, histtype='bar', ec='black') # строим гистограмму по выборке выборочных средних
plt.title('Гистограмма распределения выборочных средних при $n=5$ и \n плотность нормального распределения с соответствующими параметрами (по ЦПТ)')

norm = sts.norm(k, np.sqrt(2*k / n)) # создаем случайную нормально распределенную величину
x = np.linspace(0, 20, 100)
plt.plot(x, norm.pdf(x))
plt.xlabel('x')
````

```python
%matplotlib inline
import matplotlib.pyplot as pyplot
import scipy.stats as stats
import numpy as np

rv = stats.expon() 
sample = rv.rvs(1000)
x = np.linspace(0,15,1000)
pdf = rv.pdf(x) 
```

```python
pyplot.plot(x, pdf, linewidth = 2.5)
pyplot.hist(sample, normed = True)
pyplot.xlabel('$x$')
pyplot.ylabel('$f(x)$')
```

```python
def generate(n): 
    means = []
    for _ in range(1000):
        sample = rv.rvs(n)
        sample_mean = np.mean(sample)
        means.append(sample_mean)
    mu = 1
    sigma = (1./n) ** 0.5
    norm_rv = stats.norm(loc=mu, scale=sigma)
    x = np.linspace(0,4,100)
    pdf = norm_rv.pdf(x)
    pyplot.plot(x, pdf, linewidth = 2.5)
    pyplot.hist(means, normed = True)
    pyplot.xlabel('$\\bar{X}_n$')
    pyplot.ylabel('$f(\\bar{X}_n)$')
    pyplot.title('n = %d' % n)
    pyplot.show()
    
for n in [3, 5, 10, 50, 200]:
    generate(n)
```

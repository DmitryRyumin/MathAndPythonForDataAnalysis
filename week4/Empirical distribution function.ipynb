```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 0
sigma = 1

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)

# сгенерируем n значений
n = 100
a = norm_rv.rvs(size=n)
#print(a)
```

Построим график функции распределения:

```python
x = np.linspace(-4,4,100)
cdf = norm_rv.cdf(x) # функция может принимать и вектор (x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

Оценим функцию распределения по выборке $a$ с помощью эмпирической функции распредления:

$F_n(x) = \dfrac{1}{n} \sum^n_i \mathbf{ind}(X_i \le x)$, где $\mathbf{ind}$ - функция индикатор

```python
fn = []
for xi in x:
    fn.append( (1. / n)*(a <= xi).tolist().count(True) )
fn = np.array(fn)
```

```python
plt.plot(x, cdf, label="Distribution function")
plt.step(x, fn, label="Empirical distribution function")
plt.legend()
plt.ylabel('$F(X)$')
plt.xlabel('$x$')
plt.title("Empirical estimation of CDF")
```

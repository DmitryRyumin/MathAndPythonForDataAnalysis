```python
from numpy.linalg import norm
```

```python
def fixed_norm(v):
    #x, y = v[0], v[1]
    return (1 - norm(v))
#fixed_norm(np.array([1, 1]))
```

```python
def fixed_norm2(y, x):
    v = np.array([x, y])
    return (1 - norm(v))
#fixed_norm2(1, 1)
```

```python
from scipy import optimize
optimize.fsolve(fixed_norm2, 0, 0.25)
```

```python
x = np.linspace(-1, 1, 1000)
y = optimize.fsolve(fixed_norm2, 0, x)
print y
```

```python
plt.plot(x, y)
z = -y
plt.plot(x, z)
plt.show()
```
